```python
"""
Emotion -> Depression-risk monitoring
Full implementation script for the assignment.

Features included:
- Load dair-ai/emotion from HuggingFace Datasets
- Risk mapping: joy/love/surprise -> low, anger/fear -> mid, sadness -> high
- Zero-shot (HuggingFace zero-shot NLI using facebook/bart-large-mnli)
- Few-shot (embedding k-NN classifier using sentence-transformers; 3-5 examples per class)
- LoRA fine-tuning using HuggingFace Transformers + PEFT on a small encoder model
- Evaluation: F1 (macro), AUROC (one-vs-rest), PR-AUC (one-vs-rest), confusion matrix
- Visualization: high-risk trend plot and rolling-window heatmap

Notes:
- This script assumes you have access to GPU for LoRA training; CPU will work but slowly.
- Install requirements: see REQUIREMENTS variable below.
- Citation / dataset page: https://huggingface.co/datasets/dair-ai/emotion

Author: ChatGPT (implementation for assignment)
"""

REQUIREMENTS = """
python>=3.8
transformers
datasets
evaluate
scikit-learn
sentence-transformers
peft
accelerate
torch
matplotlib
pandas
seaborn
numpy
"""

import os
import random
from typing import List, Tuple, Dict

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import torch
from datasets import load_dataset
from sklearn.metrics import (f1_score, roc_auc_score, precision_recall_curve,
                             auc, confusion_matrix)

# ---------------------------
# 1. Load dataset & preprocessing
# ---------------------------

def load_emotion_dataset():
    """Load dair-ai/emotion from HuggingFace."""
    ds = load_dataset("dair-ai/emotion")
    # Convert to pandas for convenience in visualization / few-shot sampling
    train_df = pd.DataFrame(ds["train"])
    val_df = pd.DataFrame(ds["validation"])
    test_df = pd.DataFrame(ds["test"])
    label_names = ds["train"].features["label"].names
    return train_df, val_df, test_df, label_names

# ---------------------------
# 2. Risk mapping
# ---------------------------
# emotion labels expected: ['sadness','joy','love','anger','fear','surprise'] (check label_names)
EMOTION_TO_RISK = {
    'joy': 0,       # low
    'love': 0,      # low
    'surprise': 0,  # low
    'anger': 1,     # mid
    'fear': 1,      # mid
    'sadness': 2    # high
}
RISK_NAME = {0: 'low_risk', 1: 'mid_risk', 2: 'high_risk'}


def map_emotion_to_risk_label(emotion_label_name: str) -> int:
    return EMOTION_TO_RISK[emotion_label_name]

# ---------------------------
# 3. Zero-shot (NLI) using HF pipeline
# ---------------------------
from transformers import pipeline

zero_shot_classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

EMOTION_LABELS = ['joy','love','surprise','anger','fear','sadness']


def zero_shot_predict(texts: List[str]) -> Tuple[np.ndarray, List[int]]:
    """Return probabilities for each emotion label (order EMOTION_LABELS) and mapped risk labels.
    Uses the HuggingFace zero-shot pipeline (NLI).
    """
    probs = []
    preds = []
    for t in texts:
        out = zero_shot_classifier(t, candidate_labels=EMOTION_LABELS, multi_label=False)
        # out['scores'] aligned with out['labels']
        label_score_map = {lab: sc for lab, sc in zip(out['labels'], out['scores'])}
        scores = [label_score_map.get(l, 0.0) for l in EMOTION_LABELS]
        probs.append(scores)
        # choose emotion with highest score
        pred_emotion = out['labels'][np.argmax(out['scores'])]
        preds.append(EMOTION_TO_RISK[pred_emotion])
    return np.array(probs), preds

# ---------------------------
# 4. Few-shot: embedding k-NN (3-5 shots per emotion)
# ---------------------------
from sentence_transformers import SentenceTransformer
from sklearn.neighbors import KNeighborsClassifier

embed_model = SentenceTransformer('all-MiniLM-L6-v2')


def build_few_shot_knn(train_df: pd.DataFrame, shots_per_class: int = 5, k: int = 3):
    """Sample shots_per_class examples per emotion, embed, fit k-NN classifier on emotion labels.
       Returns classifier and the sampled examples for reference.
    """
    sampled = []
    for emo in EMOTION_LABELS:
        sub = train_df[train_df['label_name'] == emo]
        sampled_sub = sub.sample(min(len(sub), shots_per_class), random_state=42)
        sampled.append(sampled_sub)
    sampled_df = pd.concat(sampled).reset_index(drop=True)
    texts = sampled_df['text'].tolist()
    labels = [EMOTION_LABELS.index(e) for e in sampled_df['label_name']]
    embeddings = embed_model.encode(texts, show_progress_bar=True)
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(embeddings, labels)
    return knn, sampled_df


def few_shot_knn_predict(knn: KNeighborsClassifier, texts: List[str]):
    emb = embed_model.encode(texts, show_progress_bar=False)
    probs = knn.predict_proba(emb)
    preds_emotion_idx = knn.predict(emb)
    preds_risk = [EMOTION_TO_RISK[EMOTION_LABELS[i]] for i in preds_emotion_idx]
    return probs, preds_risk

# ---------------------------
# 5. LoRA fine-tuning on emotion -> we'll fine-tune an encoder model for emotion classification, then map to risk
# ---------------------------
# Uses peft.LoraConfig to add LoRA adapters to an encoder model (e.g., distilbert-base-uncased)
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
from datasets import Dataset
import evaluate
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training


def prepare_datasets_for_training(train_df: pd.DataFrame, val_df: pd.DataFrame, label_names: List[str], max_length: int = 128):
    tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')
    def tokenize(batch):
        return tokenizer(batch['text'], truncation=True, padding='max_length', max_length=max_length)
    train_ds = Dataset.from_pandas(train_df[['text','label']])
    val_ds = Dataset.from_pandas(val_df[['text','label']])
    train_ds = train_ds.map(lambda x: tokenizer(x['text'], truncation=True, padding='max_length', max_length=max_length), batched=True)
    val_ds = val_ds.map(lambda x: tokenizer(x['text'], truncation=True, padding='max_length', max_length=max_length), batched=True)
    train_ds.set_format(type='torch', columns=['input_ids','attention_mask','label'])
    val_ds.set_format(type='torch', columns=['input_ids','attention_mask','label'])
    return tokenizer, train_ds, val_ds


def train_lora_classifier(train_ds, val_ds, num_labels: int = 6, output_dir: str = './lora_emotion'):
    model_name = 'distilbert-base-uncased'
    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=num_labels)

    # Prepare model for LoRA; for encoder models LoRA still works on attention/query/key/value
    peft_config = LoraConfig(
        r=8,
        lora_alpha=32,
        target_modules=['q_lin','v_lin','k_lin'] , # distilbert naming may differ; fallback below
        lora_dropout=0.1,
        bias='none',
        task_type='SEQ_CLS'
    )
    # Since target_modules may not match distilbert, catch exceptions and use default target_modules
    try:
        model = get_peft_model(model, peft_config)
    except Exception as e:
        peft_config.target_modules = ['query','value','key','dense']
        model = get_peft_model(model, peft_config)

    training_args = TrainingArguments(
      output_dir=output_dir,
      per_device_train_batch_size=32,
      per_device_eval_batch_size=32,
      eval_strategy="epoch",           # ✅ 新版參數
      save_strategy="epoch",
      logging_strategy="epoch",
      num_train_epochs=3,
      load_best_model_at_end=True,
      metric_for_best_model="accuracy",
      report_to="none"
    )


    # Metrics
    metric_f1 = evaluate.load('f1')
    metric_acc = evaluate.load('accuracy')

    def compute_metrics(eval_pred):
        logits, labels = eval_pred
        preds = np.argmax(logits, axis=-1)
        f1 = metric_f1.compute(predictions=preds, references=labels, average='macro')
        acc = metric_acc.compute(predictions=preds, references=labels)
        return {**f1, **acc}

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_ds,
        eval_dataset=val_ds,
        compute_metrics=compute_metrics
    )

    trainer.train()
    trainer.save_model(output_dir)
    return trainer, model


def lora_predict(model, tokenizer, texts: List[str], device='cpu'):
    model.to(device)
    enc = tokenizer(texts, truncation=True, padding=True, return_tensors='pt')
    enc = {k:v.to(device) for k,v in enc.items()}
    with torch.no_grad():
        out = model(**enc)
        logits = out.logits.cpu().numpy()
        probs = torch.softmax(torch.tensor(logits), dim=-1).numpy()
        preds_idx = np.argmax(logits, axis=-1)
        preds_risk = [EMOTION_TO_RISK[EMOTION_LABELS[i]] for i in preds_idx]
    return probs, preds_risk

# ---------------------------
# 6. Evaluation utilities
# ---------------------------

def evaluate_risk_preds(y_true_risk: List[int], y_pred_risk: List[int], y_score_highrisk: List[float]):
    """Compute F1 (macro) on risk labels, AUROC and PR-AUC for high-risk class (one-vs-rest)
    y_score_highrisk should be probability of high risk (class=2)
    """
    f1_macro = f1_score(y_true_risk, y_pred_risk, average='macro')
    # AUROC for high risk
    try:
        auroc_high = roc_auc_score([1 if y==2 else 0 for y in y_true_risk], y_score_highrisk)
    except Exception:
        auroc_high = float('nan')
    # PR-AUC
    try:
        precision, recall, _ = precision_recall_curve([1 if y==2 else 0 for y in y_true_risk], y_score_highrisk)
        pr_auc_high = auc(recall, precision)
    except Exception:
        pr_auc_high = float('nan')
    cm = confusion_matrix(y_true_risk, y_pred_risk, labels=[0,1,2])
    return {'f1_macro': f1_macro, 'auroc_high': auroc_high, 'pr_auc_high': pr_auc_high, 'confusion_matrix': cm}

# ---------------------------
# 7. Visualization
# ---------------------------

def plot_high_risk_trend(probs_high: List[float], title: str = 'High-risk trend (P(high))'):
    plt.figure(figsize=(12,4))
    plt.plot(probs_high, marker='.', linewidth=1)
    plt.xlabel('Sample index')
    plt.ylabel('P(high_risk)')
    plt.title(title)
    plt.grid(True)
    plt.tight_layout()
    plt.show()


def plot_high_risk_heatmap(probs_high: List[float], window: int = 50, title: str = 'High-risk rolling heatmap'):
    # Compute rolling average with stride and show as heatmap
    arr = np.array(probs_high)
    n = len(arr)
    # create matrix: each row is a window average over window size sliding by stride
    stride = max(1, window//4)
    rows = []
    for start in range(0, n-window+1, stride):
        rows.append(arr[start:start+window].mean())
    mat = np.array(rows).reshape(-1,1)
    plt.figure(figsize=(6,8))
    sns.heatmap(mat, cbar=True, yticklabels=False)
    plt.title(title)
    plt.xlabel('Rolling window average P(high)')
    plt.ylabel('Window index')
    plt.tight_layout()
    plt.show()

# ---------------------------
# 8. Main run (glue everything)
# ---------------------------

def main():
    # 1) load
    train_df, val_df, test_df, label_names = load_emotion_dataset()
    print('Label names detected:', label_names)
    # Ensure label_name column exists (some hf dataset provide numeric 'label')
    if 'label_name' not in train_df.columns:
        # map numeric label to names
        train_df['label_name'] = train_df['label'].map(lambda x: label_names[x])
        val_df['label_name'] = val_df['label'].map(lambda x: label_names[x])
        test_df['label_name'] = test_df['label'].map(lambda x: label_names[x])

    # 2) Zero-shot on test set (fast)
    print('Running zero-shot on test split...')
    zs_probs, zs_preds_risk = zero_shot_predict(test_df['text'].tolist())
    # compute probability of high-risk (sadness) by summing scores for sadness label only
    # ensure alignment: EMOTION_LABELS == order returned earlier; zero_shot_predict uses EMOTION_LABELS order
    idx_sadness = EMOTION_LABELS.index('sadness')
    zs_p_high = zs_probs[:, idx_sadness].tolist()
    zs_eval = evaluate_risk_preds(test_df['label'].map(lambda x: EMOTION_TO_RISK[label_names[x]]).tolist(), zs_preds_risk, zs_p_high)
    print('\nZero-shot results:', zs_eval)

    # 3) Few-shot KNN
    print('Building few-shot KNN (5 shots per class)...')
    knn, shots_df = build_few_shot_knn(train_df, shots_per_class=5, k=3)
    fs_probs, fs_preds_risk = few_shot_knn_predict(knn, test_df['text'].tolist())
    # identify index of sadness
    # Note: knn classes correspond to index in EMOTION_LABELS (we used that mapping earlier)
    idx_sad = EMOTION_LABELS.index('sadness')
    if fs_probs.shape[1] == len(EMOTION_LABELS):
        fs_p_high = fs_probs[:, idx_sad].tolist()
    else:
        # fallback: use predicted label probability as 1 for predicted label
        fs_p_high = [p[idx_sad] if idx_sad < p.shape[0] else 0.0 for p in fs_probs]
    fs_eval = evaluate_risk_preds(test_df['label'].map(lambda x: EMOTION_TO_RISK[label_names[x]]).tolist(), fs_preds_risk, fs_p_high)
    print('\nFew-shot KNN results:', fs_eval)

    # 4) LoRA training (this can be slow; configured for small epoch count)
    print('Preparing datasets for LoRA training...')
    tokenizer, train_ds, val_ds = prepare_datasets_for_training(train_df, val_df, label_names, max_length=128)
    print('Starting LoRA training (distilbert-base-uncased)')
    trainer, lora_model = train_lora_classifier(train_ds, val_ds, num_labels=len(label_names))
    # Predict with LoRA model on test set
    print('Predicting with LoRA-finetuned model...')
    test_texts = test_df['text'].tolist()
    lora_probs, lora_preds_risk = lora_predict(lora_model, tokenizer, test_texts, device='cuda' if torch.cuda.is_available() else 'cpu')
    # lora_probs: probabilities over emotions following EMOTION_LABELS order used in training from dataset.label numeric mapping
    # We must ensure mapping between model label order and EMOTION_LABELS. We used dataset numeric label mapping earlier; label_names give correct order.
    # index of sadness in label_names
    idx_sad_lora = label_names.index('sadness')
    lora_p_high = lora_probs[:, idx_sad_lora].tolist()
    lora_eval = evaluate_risk_preds(test_df['label'].map(lambda x: EMOTION_TO_RISK[label_names[x]]).tolist(), lora_preds_risk, lora_p_high)
    print('\nLoRA fine-tuned results:', lora_eval)

    # 5) Visualizations (use lora_p_high as primary if available; otherwise use zero-shot)
    probs_high_ref = lora_p_high if len(lora_p_high) == len(test_texts) else zs_p_high
    plot_high_risk_trend(probs_high_ref, title='High-risk trend (LoRA model P(high))')
    plot_high_risk_heatmap(probs_high_ref, window=100, title='High-risk rolling heatmap (LoRA)')

    # Done
    print('All steps completed.')


if __name__ == '__main__':
    main()
```
